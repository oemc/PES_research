```python
import requests

indicators = [
'DT.ODA.ALLD.CD',
'DT.ODA.ODAT.PC.ZS',
'CC.EST',
'GE.EST',
'PV.EST',
'RQ.EST',
'RL.EST',
'VA.EST',
'NY.GDP.PCAP.CD',
'SP.POP.GROW'
]
for indicator in indicators:
    rows = []
    countries = 'AW;AL;AD;AE;AR;AM;AS;AG;AU;AT;AZ;BE;BG;BH;BS;BA;BY;BZ;BM;BR;BB;BN;BW;CA;CH;JG;CL;CN;CO;CR;CU;CW;KY;CY;CZ;DE;DM;DK;DO;DZ;EC;ES;EE;FI;FJ;FR;FO;GA;GB;GE;GI;GQ;GR;GD;GL;GT;GU;GY;HK;HR;HU;ID;IM;IE;IR;IQ;IS;IL;IT;JM;JP;KZ;KN;KR;KW;LY;LC;LI;LT;LU;LV;MO;MF;MC;MD;MV;MX;MH;MK;MT;ME;MN;MP;MU;MY;NA;NC;NL;NO;NR;NZ;OM;PA;PE;PW;PL;PR;PT;PY;PF;QA;RO;RU;SA;SG;SV;SM;RS;SR;SK;SI;SE;SX;SC;TC;TH;TM;TO;TT;TR;TV;TW;UA;UY;US;VC;VG;VI;XK;ZA'
    url = f'https://api.worldbank.org/v2/en/country/{countries}/indicator/{indicator}?format=json&date=1995:2023&per_page=50'

    response = requests.get(url)

    if response.status_code == 200:  # Verifica que la solicitud fue exitosa
        data = response.json()       # Convierte la respuesta JSON a un diccionario de Python
        print(data[0])
        for row in data[1]:
            rows.append(f"{row['indicator']['id']},{row['country']['id']},{row['date']},{row['value']}")
    else:
        print(f"Error {response.status_code}: No se pudo obtener los datos.")



    for i in range(2, data[0]['pages'] + 1):
        print(f'Se hace request GET {i} para el indicador {indicator}')
        
        response = requests.get(f'{url}&page={i}')
        if response.status_code == 200:  # Verifica que la solicitud fue exitosa
            data = response.json()       # Convierte la respuesta JSON a un diccionario de Python
            for row in data[1]:
                rows.append(f"{row['indicator']['id']},{row['country']['id']},{row['date']},{row['value']}")
        else:
            print(f"Error {response.status_code}: No se pudo obtener los datos.")
            break

    # Abrir un archivo en modo escritura
    with open(f"datos_python{indicator.replace('.','_')}.csv", 'w') as file:
        for item in rows:
            file.write(f"{item}\n")  # Escribir cada elemento en una nueva lnea
```
