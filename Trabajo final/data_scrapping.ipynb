```python
import requests

indicators = [
('DT.ODA.ODAT.GN.ZS', 'ODA.GNI'),
('DT.ODA.ALLD.CD', 'ODA.ALL'),
('DT.ODA.ODAT.PC.ZS', 'ODA.PC'),
('CC.EST', 'CC'),
('GE.EST', 'GE'),
('PV.EST', 'PV'),
('RQ.EST', 'RQ'),
('RL.EST', 'RL'),
('VA.EST', 'VA'),
('NY.GDP.PCAP.CD', 'GDP.PC'),
('NY.GDP.PCAP.PP.CD', 'GDP.PPP'),
('SP.POP.GROW', 'GROW'),
('NY.GDS.TOTL.ZS', 'SAV')
]
for (indicator, name) in indicators:
    rows = []
    countries = 'AW;AF;AO;AL;AD;AE;AR;AM;AS;AG;AU;AT;AZ;BI;BE;BJ;BF;BD;BG;BH;BS;BA;BY;BZ;BM;BO;BR;BB;BN;BT;BW;CF;CA;CH;JG;CL;CN;CI;CM;CD;CG;CO;KM;CV;CR;CU;CW;KY;CY;CZ;DE;DJ;DM;DK;DO;DZ;EC;EG;ER;ES;EE;ET;FI;FJ;FR;FO;FM;GA;GB;GE;GH;GI;GN;GM;GW;GQ;GR;GD;GL;GT;GU;GY;HK;HN;HR;HT;HU;ID;IM;IN;IE;IR;IQ;IS;IL;IT;JM;JO;JP;KZ;KE;KG;KH;KI;KN;KR;KW;LA;LB;LR;LY;LC;LI;LK;LS;LT;LU;LV;MO;MF;MA;MC;MD;MG;MV;MX;MH;MK;ML;MT;MM;ME;MN;MP;MZ;MR;MU;MW;MY;NA;NC;NE;NG;NI;NL;NO;NP;NR;NZ;OM;PK;PA;PE;PH;PW;PG;PL;PR;KP;PT;PY;PS;PF;QA;RO;RU;RW;SA;SD;SN;SG;SB;SL;SV;SM;SO;RS;SS;ST;SR;SK;SI;SE;SZ;SX;SC;SY;TC;TD;TG;TH;TJ;TM;TL;TO;TT;TN;TR;TV;TW;TZ;UG;UA;UY;US;UZ;VC;VG;VI;VN;VU;WS;XK;YE;ZA;ZM;ZW'
    url = f'https://api.worldbank.org/v2/en/country/{countries}/indicator/{indicator}?format=json&date=1990:2023&per_page=50'

    response = requests.get(url)

    if response.status_code == 200:  # Verifica que la solicitud fue exitosa
        data = response.json()       # Convierte la respuesta JSON a un diccionario de Python
        print(data[0])
        for row in data[1]:
            rows.append(f"{name},{row['country']['id']},{row['date']},{row['value']}")
    else:
        print(f"Error {response.status_code}: No se pudo obtener los datos.")



    for i in range(2, data[0]['pages'] + 1):
        print(f'Se hace request GET {i} para el indicador {indicator}')
        
        response = requests.get(f'{url}&page={i}')
        if response.status_code == 200:  # Verifica que la solicitud fue exitosa
            data = response.json()       # Convierte la respuesta JSON a un diccionario de Python
            for row in data[1]:
                rows.append(f"{name},{row['country']['id']},{row['date']},{row['value']}")
        else:
            print(f"Error {response.status_code}: No se pudo obtener los datos.")
            break

    # Abrir un archivo en modo escritura
    with open(f"descarga/datos_python_{name}.csv", 'w') as file:
        for item in rows:
            file.write(f"{item}\n")  # Escribir cada elemento en una nueva lnea
```
